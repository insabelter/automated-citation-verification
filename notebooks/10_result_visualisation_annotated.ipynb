### Distribution of Citation Functions

| **Main Category** | **Sub Category**            | **Explanation**                                                                                     |
|--------------------|-----------------------------|-----------------------------------------------------------------------------------------------------|
| Use               | Apply                       | The methods/algorithms/tools/data are directly applied and not modified.                           |
|                   | Extend                      | The methods/algorithms/tools are modified before/while applying them, meaning the reference work is only used as a basis to build upon. |
| Compare           | Criticize                   | Negative comparison: Reference is criticized, weaknesses are pointed out, errors are identified, citing author does not agree with reference. |
|                   | Contrast                    | Neutral comparison: Comparison without clear positive or negative sentiment of the citing author in relation to the reference(s). |
|                   | Confirm                     | Positive comparison: Reference results are supported by the author or can be confirmed/validated by the citing work’s results. |
| Related           | Definition/Proof            | Definitions are directly taken from the reference, or result data is directly taken from the reference – or they are referenced to provide proof for a claim. |
|                   | Fundamentals                | The reference provides fundamental knowledge to understand the citing author’s work.               |
|                   | Acknowledge                 | The reference is related to the citing author’s work but not needed to understand it. This includes examples or mentioning pioneers. |
| Background        | Introduction/Bigger picture | The reference is cited to give an introduction to the topic or to locate the citing author’s work within the broader landscape of the field, even though the reference topic is not directly related to the citing paper. |
|                   | Unrelated/Unclear           | The reference topic is unrelated to the citing paper or it is unclear which category else to pick because there is not enough context given. |

### Citation Function by Section

```python
citation_functions = {
    "Use": ["Apply", "Extend"],
    "Compare": ["Criticize", "Contrast", "Confirm"],
    "Related": ["Definition/Proof", "Fundamentals", "Acknowledge"],
    "Background": ["Introduction/Bigger picture", "Unrelated/Unclear"]
}
main_functions = list(citation_functions.keys())
sub_functions = [sub for sub_list in citation_functions.values() for sub in sub_list]

statement_sections = ["Introduction", "Background", "Related Work", "Methods", "Results", "Discussion"]
```

```python
# Create a copy of df_annotated and clean up the Citation Function: Main column
df_sorted = df_annotated.copy()
df_sorted["Citation Function: Main"] = df_sorted["Citation Function: Main"].str.strip()  # Remove whitespace
df_sorted["Citation Statement Section"] = pd.Categorical(
    df_sorted["Citation Statement Section"], categories=statement_sections, ordered=True
)

# Group and sort the rows and columns by the defined orders
counts = df_sorted.groupby(
    ["Citation Statement Section", "Citation Function: Main"], observed=False
).size().unstack(fill_value=0)
counts = counts.reindex(index=statement_sections, columns=main_functions)  # Sort rows and columns
display(counts)
```

```python
# Create a copy of df_annotated and clean up the Citation Function: Sub column
df_sorted = df_annotated.copy()
df_sorted["Citation Function: Sub"] = df_sorted["Citation Function: Sub"].str.strip()  # Remove whitespace
df_sorted["Citation Statement Section"] = pd.Categorical(
    df_sorted["Citation Statement Section"], categories=statement_sections, ordered=True
)

# Group and sort the rows and columns by the defined orders
counts = df_sorted.groupby(
    ["Citation Statement Section", "Citation Function: Sub"], observed=False
).size().unstack(fill_value=0)
counts = counts.reindex(index=statement_sections, columns=sub_functions)  # Sort rows and columns
display(counts)
```

You can see that the citation statement section and the citation function do not seem completely independet of each other. For example, most of the "Introduction/Bigger picture" citations appear in the "Introduction" section of the paper and "Apply" citations are mostly found in the "Method" section.

As in the annotated dataset there are now only two labels used: fully substantiated and unsubstantiated, the accuracy when prompting for three labels (including partially substantiated) goes down as the correct classifications for the previously as partially substanitated labelled statements are removed and this is not replaced with the amount of now correct classified fully/unsubstantiated statements. 

#### Accuracy of Re-Annotated Data vs. same rows of Original Data (prompting for three labels)

```python
df_annotated = dfs[dfs_info['df_annot']['id']]
df_annotated = df_annotated[df_annotated["Citation Statement Section"].notna()]

df_prev = dfs[dfs_info['df_prev']['id']]
df_prev_filtered = df_prev[df_prev.index.isin(df_annotated.index)]

assert all(df_annotated["Citing Article ID"] == df_prev_filtered["Citing Article ID"])
assert all(df_annotated["Reference Article ID"] == df_prev_filtered["Reference Article ID"])

results_annotated = eval_predictions_all_labels(df_annotated)
results_prev_filtered = eval_predictions_all_labels(df_prev_filtered)

compare_results = {
    "original_data_same_rows": {
        "all_labels": results_prev_filtered
    },
    "annotated_data_rows": {
        "all_labels": results_annotated
    },
}

print("Portion of correctly classified per label:")
print_table_label_accuracies(calc_label_accuracies(compare_results, as_portions=True), string_given=True)

print("\nAccuracy per label:")
print_table_label_accuracies(calc_label_accuracies(compare_results))
```

It can be seen that most of the rows which were labelled as "partially substantiated" in the original dataset and are now labelled as "fully substantiated" in the re-annotated dataset are now false classified by the model, probably as "partially substantiated". The accuracy on the re-annotated dataset rows is therefore lower than the accuracy on the original dataset rows as the model did not know in this round that the "partially substantiated" label was not used anymore.
